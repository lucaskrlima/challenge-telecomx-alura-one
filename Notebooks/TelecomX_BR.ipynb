```python
# Importando bibliotecas padrÃ£o
import requests
import json
import pandas as pd
```

```python
# Carregando os dados da API
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json'
dados_telecomx = requests.get(url)

# Adicionando os dados em uma variÃ¡vel
dados = json.loads(dados_telecomx.text)
```

```python
# Exibindo os dados
pd.DataFrame(dados)
```

```python
# Convertendo e normalizando os dados em um dataframe
dados_normalizado = pd.json_normalize(dados, sep='_')

dados_normalizado
```

## *Verificando InconsistÃªncias nos Dados*

```python
#Verificando existÃªncia de dados nulos
dados_normalizado.isnull().sum()
```

```python
# Convertendo a coluna account_Charges_Total para float
dados_normalizado['account_Charges_Total'] = pd.to_numeric(dados_normalizado['account_Charges_Total'], errors='coerce')
```

```python
#Tratandos algumas inconsistÃªncias deixando todos os nomes das colunas em minÃºsculos
dados_normalizado.columns = dados_normalizado.columns.str.lower()
```

```python
#Tratando os dados de customer id, retirando o -
dados_normalizado.customerid.unique()
```

```python
dados_normalizado['customerid'] = dados_normalizado['customerid'].str.replace('-', '')
```

```python
dados_normalizado.customerid.unique()
```

```python
#Renomeando os nomes das colunas
nome_colunas_atualizadas = {

    'customerid': 'id_cliente',
    'customer_seniorcitizen': 'cidadao_senior',
    'customer_partner': 'parceiro',
    'customer_dependents': 'dependentes',
    'customer_tenure': 'meses_contrato',
    'phone_phoneservice': 'servico_telefone',
    'phone_multiplelines': 'multiplas_linhas',
    'internet_internetservice': 'servico_internet',
    'internet_onlinesecurity': 'seguranca_online',
    'internet_onlineBackup': 'backup_online',
    'internet_deviceoprotection': 'protecao_dispositivo',
    'internet_techsupport': 'suporte_tecnico',
    'internet_streamingtv': 'streaming_tv',
    'internet_StreamingMovies': 'streaming_filmes',
    'account_contract': 'contrato',
    'account_paperlessbilling': 'fatura_online',
    'account_paymentmethod': 'metodo_pagamento',
    'account_charges_monthly': 'gastos_mensais',
    'account_charges_total': 'gastos_totais'
}
```

```python
dados_normalizado.rename(columns=nome_colunas_atualizadas, inplace=True)
dados_normalizado.columns.unique()
```

```python
#Cria uma coluna conta_diaria que contÃ©m o gasto diÃ¡rio com base em gastos_mensais
dados_normalizado['contas_diarias'] = dados_normalizado['gastos_mensais'] / 30
```

## *AnÃ¡lise* *Descritiva*

```python
#Calculando mÃ©dia, mediana e desvio padrÃ£o com mÃ©todo describe
dados_normalizado.describe()
```

```python
#DistribuiÃ§Ã£o de evasÃ£o entre clientes e servico_internet
import seaborn as sns
import matplotlib.pyplot as plt

#Criando um grÃ¡fico de contagem para distribuiÃ§Ã£o de churn
sns.countplot(x='servico_internet', hue='churn', data=dados_normalizado, palette='viridis')

#TÃ­tulos e rÃ³tulos
plt.title('DistribuiÃ§Ã£o de churn por ServiÃ§o de internet')
plt.xlabel('ServiÃ§o de internet')
plt.ylabel('Contagem de clientes')
plt.legend(title='Churn', labels=['NÃ£o', 'Sim'])

plt.show()
```

```python
#DistribuiÃ§Ã£od e gastos mensais por churn

#Cria um box plot para Gastos mensais x Churn
sns.boxplot(x='churn', y='gastos_mensais', data=dados_normalizado, palette='pastel')

plt.title('DistribuiÃ§Ã£o de gastos mensais por churn')
plt.xlabel('Churn')
plt.ylabel('Gastos mensais')

plt.show()
```

```python
#Contagem de evasÃ£o por uma variÃ¡vel categÃ³rica usando metodo de tabela cruzada
proporcao_por_genero = pd.crosstab(dados_normalizado['customer_gender'], dados_normalizado['churn'], normalize='index')
print(proporcao_por_genero)
```

```python
'''
Contagem e evasÃ£o por variÃ¡veis numÃ©ricas

Boxplot de distribuiÃ§Ã£o de gastos totais por evasÃ£o
'''
#Definindo o tamanho do grÃ¡fico
plt.figure(figsize=(10,6))

#Criando boxplot
sns.boxplot(x='churn', y='gastos_mensais', data=dados_normalizado, palette='pastel')

#RÃ³tulos e tÃ­tulos
plt.title('DistribuiÃ§Ã£o de gastos totais por evasÃ£o(churn)')
plt.xlabel('EvasÃ£o')
plt.ylabel('Gastos totais')

plt.show()
```

```python
#Definindo o tamanho do grÃ¡fico
plt.figure(figsize=(10,6))

#Criando grÃ¡fico de densidade
sns.kdeplot(x='meses_contrato', hue='churn', data=dados_normalizado, fill=True, common_norm=False)

#RÃ³tulos e tÃ­tulos
plt.title('DistribuiÃ§Ã£o de meses de contrato por evasÃ£o(churn)')
plt.xlabel('Meses de contrato')
plt.ylabel('Densidade')

plt.show()
```
